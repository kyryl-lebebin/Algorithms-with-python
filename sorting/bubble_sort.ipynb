```python
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n-i-1):
            # Traverse the array from 0 to n-i-1. Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

    return arr
```

#### Time complexity

- at the ith iteration on outer loop, all the elements at positions i...n are sorted
- terminate if no swaps executed during inner for loop

**Worst case**: O(n^2): (n-1)+(n-2)+...+1=$\frac{n(n-1)}{2}$ - reversly sorted list
**Best case**: O(n): n-1 - initially sorted list
